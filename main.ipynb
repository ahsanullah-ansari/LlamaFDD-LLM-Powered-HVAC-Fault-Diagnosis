```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# Make sure you're using GPU
print("Using device:", torch.cuda.get_device_name(0) if torch.cuda.is_available() else "CPU")

# Load tokenizer (requires login if model is gated/private)
tokenizer = AutoTokenizer.from_pretrained(
    "meta-llama/Llama-3.2-3B-Instruct",
    use_auth_token=True  # Needed for Llama models
)

# Load model and send to GPU (automatically picks GPU if available)
model = AutoModelForCausalLM.from_pretrained(
    "meta-llama/Llama-3.2-3B-Instruct",
    device_map="auto",
    torch_dtype=torch.float16,  # Use float16 for faster inference on GPU
    use_auth_token=True
)

# Sample prompt
prompt = "Explain the difference between supervised and unsupervised learning."

# Tokenize and move input to the same device as model
inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

# Generate output
outputs = model.generate(
    **inputs,
    max_new_tokens=150,
    do_sample=True,
    temperature=0.7,
    top_p=0.9
)

# Decode and print
print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```
